Q.No-01    What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans :-

Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations of data. It provides a wide range of plotting capabilities and is extensively used in data analysis, scientific research, and various fields of data visualization. Matplotlib is highly customizable, allowing users to create high-quality plots with various styles, colors, labels, and annotations.

````python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 6, 3]
````

Here are five types of plots that can be created using the Pyplot module of Matplotlib:

1. Line Plot: A line plot is used to visualize the relationship between two numerical variables. It consists of data points connected by straight lines, often used to show trends or patterns over time or continuous values.

````python
# Create a line plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')

# Display the plot
plt.show()
````

2. Scatter Plot: Scatter plots are used to represent the relationship between two variables, showing individual data points as dots on the graph. This type of plot is useful for identifying patterns, clusters, or correlations in the data.

````python
# Create a scatter plot
plt.scatter(x, y, color='r', marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Display the plot
plt.show()
````

3. Bar Plot: Bar plots, also known as bar charts, are used to display categorical data with rectangular bars. Each bar's height or length corresponds to the frequency or value of the category it represents, making it easy to compare different categories.

````python
# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [10, 15, 7, 20]

# Create a bar plot
plt.bar(categories, values, color='g')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

# Display the plot
plt.show()
````

4. Histogram: A histogram is used to visualize the distribution of a single numerical variable. It groups the data into bins and displays the frequency of data points falling into each bin, giving an overview of the data's underlying distribution.

````python
# Sample data
data = [1, 1, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 7]

# Create a histogram
plt.hist(data, bins=7, color='purple', edgecolor='black')

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Display the plot
plt.show()
````

5. Pie Chart: Pie charts are used to represent the composition of a whole by dividing it into slices. Each slice's size represents the proportion of the whole that belongs to a specific category, making it easy to understand the relative contributions of different parts.

````python
# Sample data
sizes = [30, 25, 15, 20, 10]

# Labels for the pie chart slices
labels = ['Slice 1', 'Slice 2', 'Slice 3', 'Slice 4', 'Slice 5']

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['orange', 'blue', 'green', 'red', 'purple'])

# Add a title
plt.title('Pie Chart Example')

# Display the plot
plt.show()
````

Q.No-02    What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```
    Note: Also add title, xlabel, and ylabel to the plot.

Ans :-

A scatter plot is a type of data visualization that represents individual data points on a graph. It is used to display the relationship between two variables and helps identify any patterns or trends in the data. Each data point is represented by a dot on the plot, with the x-coordinate representing one variable and the y-coordinate representing the other variable.

````python
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, marker='o', color='b', label='Data Points')

# Add title and labels
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.legend()
plt.grid(True)
plt.show()
````

Explanation:

1. We imported the necessary libraries: numpy for generating random data and matplotlib.pyplot for plotting the scatter plot.
2. We used the `np.random.normal` function to generate 50 random data points for 'x' with a mean of 3 and standard deviation of 2. Similarly, we generated 'y' with the same mean and standard deviation as 'x'.
3. We then used `plt.scatter()` to create the scatter plot, passing 'x' and 'y' as input, with marker 'o' (circle) and color 'b' (blue).
4. `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` are used to add the title, x-axis label, and y-axis label to the plot, respectively.
5. `plt.legend()` is used to display the legend with the label 'Data Points'.
6. Finally, we display the plot using `plt.show()`.

Q.No-03    Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
```python
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

Ans :-

The `subplot()` function is used in Matplotlib, a popular data visualization library in Python, to create multiple plots within the same figure. It allows you to organize multiple plots in a grid-like fashion, making it easier to compare and analyze different data visualizations side by side.

The syntax for `subplot()` is as follows:
```python
    subplot(nrows, ncols, index)
```

where `nrows` and `ncols` specify the number of rows and columns of the subplot grid, and `index` indicates the position of the current plot in the grid, starting from 1 (top-left corner) to `nrows * ncols` (bottom-right corner).

Now, let's create four line plots using the `subplot()` function:

    For Line 1

````python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='Line 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line 1')
````

    For Line 2

````python
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2, 2, 2)
plt.plot(x2, y2, label='Line 2', color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line 2')
````

    For Line 3

````python
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2, 2, 3)
plt.plot(x3, y3, label='Line 3', color='green')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line 3')
````

    For Line 4

````python
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 4)
plt.plot(x4, y4, label='Line 4', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line 4')
````

Q.No-04    What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 
```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

Ans :-

A bar plot, also known as a bar chart, is a type of data visualization that represents categorical data with rectangular bars. Each bar typically corresponds to a category, and the length or height of the bar represents the value associated with that category. Bar plots are used to compare and display the relative values of different categories and are particularly useful for visualizing discrete data.

Bar plots are commonly used to :

1. `Compare data` : Bar plots make it easy to compare values between different categories. This visual comparison helps identify trends, patterns, and outliers.

2. `Show distribution` : Bar plots can also display the distribution of data across various categories, allowing viewers to understand the frequency of occurrence in each category.

3. `Present rankings` : When the data is sorted based on some criteria, bar plots can show the ranking of different categories based on their values.

Now, let's create a bar plot and a horizontal bar plot using the given data:

````python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
````

````python
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.title("Profit Comparison of Companies")
plt.show()
````

````python
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.title("Profit Comparison of Companies")
plt.show()
````

Explanation:

1. We use `plt.bar()` to create a vertical bar plot. The `company` array is used as the x-axis (category labels), and the `profit` array is used as the height of the bars.

2. We use `plt.barh()` to create a horizontal bar plot. The `company` array is used as the y-axis (category labels), and the `profit` array is used as the width of the bars.

3. We set the axis labels and the plot title using `plt.xlabel()`, `plt.ylabel()`, and `plt.title()`.

4. We finally call `plt.show()` to display the plots.

Q.No-05    What is a box plot? Why is it used? Using the following data plot a box plot.
```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

Ans :-

A box plot, also known as a box-and-whisker plot, is a graphical representation of data that provides a quick summary of the distribution of a dataset. It displays the data's median, quartiles, and potential outliers. The plot consists of a rectangular box, which represents the interquartile range (IQR) of the data, and "whiskers" extending from the box to the minimum and maximum values within a certain range. Any data points beyond the whiskers are considered outliers.

The box plot is useful for several reasons:

1. **Visualizing Data Distribution**: Box plots allow us to see the spread and symmetry of data at a glance, making it easy to compare multiple datasets or variables.

2. **Identification of Outliers**: Box plots can identify potential outliers, which are data points that fall significantly outside the overall pattern of the data.

3. **Comparison of Distributions**: By placing multiple box plots side by side, we can compare the distributions of different datasets.

4. **Data Skewness**: Box plots can reveal whether data is skewed to one side or is symmetrically distributed.

Create a box plot using the provided data `box1` and `box2`

````python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data_to_plot = [box1, box2]

plt.figure(figsize=(8, 6))
plt.boxplot(data_to_plot, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot of Box 1 and Box 2')
plt.grid(True)
plt.show()
````
